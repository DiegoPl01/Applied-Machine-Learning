```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
```

```python
# 10
from keras import models,layers
```

```python
# 11
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
myNetwork = Sequential()
```

```python
# 12 method 1 from book
train_images_reshaped = train_images.reshape(len(train_images), 28*28)
test_images_reshaped = test_images.reshape(len(test_images), 28*28)
```

```python
# 12 from 20 minutes video
myNetwork.add(Flatten())
```

```python
# 13 method 1 from book
train_images_reshaped = train_images_reshaped.astype("float32") / 255
test_images_reshaped = test_images_reshaped.astype("float32") / 255
```

```python
# 13 from 20 minutes video
train_images = tensorflow.keras.utils.normalize(train_images, axis = 1)
test_images = tensorflow.keras.utils.normalize(test_images, axis = 1)
```

```python
# 14
myNetwork.add(Dense(512, activation="relu", input_dim= 784))
```

```python
# 15
myNetwork.add(Dense(128, activation="relu"))
```

```python
# 16
myNetwork.add(Dense(10, activation="softmax"))
```

```python
# 17 for compile part 1
myNetwork.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
# 18 for compile part 1
myNetwork.fit(train_images_reshaped, train_labels, epochs= 10)
```

```python
# 20 for compile part 1
myNetwork.evaluate(test_images_reshaped, test_labels)
```

```python
# 17 for compile part 2
myNetwork.compile(
    optimizer="rmsprop",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
train_labels_cat = to_categorical(train_labels)
test_labels_cat = to_categorical(test_labels)
```

```python
# 18 for compile part 2
myNetwork.fit(train_images_reshaped, train_labels_cat, epochs= 10)
```
